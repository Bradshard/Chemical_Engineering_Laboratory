```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_mixing(N, M):
    total_molecules = N + M
    num_A_in_one_side = np.random.binomial(total_molecules, N / total_molecules)
    return num_A_in_one_side

def plot_distribution(N, M, num_simulations=1000):
    results = [simulate_mixing(N, M) for _ in range(num_simulations)]
    plt.hist(results, bins=np.arange(N+2), density=True, alpha=0.75)
    plt.xlabel('Number of Type A Molecules')
    plt.ylabel('Probability')
    plt.title('Distribution of Type A Molecules After Mixing')
    plt.show()

    mean = np.mean(results)
    std_dev = np.std(results)
    print(f"Mean: {mean}")
    print(f"Standard Deviation: {std_dev}")

# Simulate mixing for N=10 and M=10
N = 10
M = 10
plot_distribution(N, M)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_mixing(N, M, num_simulations=1000):
    results = []
    for _ in range(num_simulations):
        N_A = np.random.randint(0, N + 1)
        results.append(N_A)
    return results

def plot_distribution(N, M, num_simulations=1000):
    results = simulate_mixing(N, M, num_simulations)
    plt.hist(results, bins=np.arange(N+2), density=True, alpha=0.75)
    plt.xlabel('Number of Type A Molecules')
    plt.ylabel('Probability')
    plt.title(f'Distribution of Type A Molecules After Mixing (N={N}, M={M})')
    plt.show()

    mean = np.mean(results)
    std_dev = np.std(results)
    print(f"Mean: {mean}")
    print(f"Standard Deviation: {std_dev}")

# Parameters
N_values = [10, 20, 50,1000]  # Different values of N
M_values = [10, 20, 50,1000]  # Different values of M
num_simulations = 10000

# Plot distributions for different N and M
for N in N_values:
    for M in M_values:
        plot_distribution(N, M, num_simulations)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_mixing_binomial(N, M, num_simulations=1000):
    results = np.random.binomial(N + M, 0.5, num_simulations)
    return results

def plot_distribution_binomial(N, M, num_simulations=1000):
    results = simulate_mixing_binomial(N, M, num_simulations)
    plt.hist(results, bins=np.arange(N+M+2), density=True, alpha=0.75)
    plt.xlabel('Number of Type A Molecules')
    plt.ylabel('Probability')
    plt.title(f'Distribution of Type A Molecules After Mixing (N={N}, M={M})')
    plt.show()

    mean = np.mean(results)
    std_dev = np.std(results)
    print(f"Mean: {mean}")
    print(f"Standard Deviation: {std_dev}")

# Parameters
N_values = [10, 20, 50]  # Different values of N
M_values = [10, 20, 50]  # Different values of M
num_simulations = 1000

# Plot distributions for different N and M
for N in N_values:
    for M in M_values:
        plot_distribution_binomial(N, M, num_simulations)
```
