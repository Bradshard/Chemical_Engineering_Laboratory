```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
delta_x = 0.1
delta_t = 0.1
t_max = 3.0  # Maximum time to simulate
x_min = -5.0  # Minimum value of x for visualization
x_max = 5.0  # Maximum value of x for visualization

# Calculate the number of grid points
num_x = int((x_max - x_min) / delta_x) + 1
num_t = int(t_max / delta_t) + 1

# Create the grid
x = np.linspace(x_min, x_max, num_x)
t = np.linspace(0.0, t_max, num_t)

# Initialize the solution matrix
u = np.zeros((num_t, num_x))

# Set the initial condition
u[0, np.logical_and(3.5 <= x, x <= 4)] = 1.0

# Iterate over time steps
for i in range(1, num_t):
    # Iterate over grid points
    for j in range(1, num_x - 1):
        # Calculate the derivative
        du_dx = (u[i-1, j+1] - u[i-1, j-1]) / (2.0 * delta_x)
        # Update the solution using finite difference approximation
        u[i, j] = u[i-1, j] + 0.5 * delta_t * du_dx - 0.4 * delta_t * u[i-1, j]

# Plot the solution
plt.imshow(u, extent=[x_min, x_max, 0, t_max], origin='lower', cmap='jet', aspect='auto')
plt.colorbar(label='u(x, t)')
plt.xlabel('x')
plt.ylabel('t')
plt.title('Solution of the PDE du/dt = 0.5*du/dx - 0.4u')
plt.show()
```
